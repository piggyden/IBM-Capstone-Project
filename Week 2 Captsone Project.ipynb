#### Data Analysis
I have categories the imported column from the file, the columns were defined below: 

1.	Severity Code - A code that corresponds to the severity of the collision: 1=Property Damage only, 2=Result in Injury
2.	Collision Type – Type of crash
3.	Speeding - Whether or not speeding was a factor in the collision. (Y/N)
4.	Light Condition - The light conditions during the collision.
5.	Road Condition - The condition of the road during the collision.
6.	Weather - A description of the weather conditions during the time of the collision.

#### Results
The data are imported and visualise in the form of bar graph.

#### Conclusion
From the result set we have computed, it concluded that speeding was not always a factor in the collision.
